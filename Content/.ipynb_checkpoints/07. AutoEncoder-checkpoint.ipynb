```python
import numpy as np
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)
train_X, train_Y, test_X, test_Y = mnist.train.images, mnist.train.labels, mnist.test.images, mnist.test.labels
```

```python
display_step = 500
number_features = train_X.shape[1] #784
number_output_classes = train_Y.shape[1]

num_hidden_1 = 256
num_hidden_2 = 128
```

```python
def weight_init(shape):
    return tf.Variable(tf.random_normal(shape))
```

```python
encoder_h1 = weight_init([number_features, num_hidden_1])
encoder_b1 = weight_init([num_hidden_1])
encoder_h2 = weight_init([num_hidden_1, num_hidden_2])
encoder_b2 = weight_init([num_hidden_2])
decoder_h1 = weight_init([num_hidden_2, num_hidden_1])
decoder_b1 = weight_init([num_hidden_1])
decoder_h2 = weight_init([num_hidden_1, number_features])
decoder_b2 = weight_init([number_features])
```

```python
def encoder(X, encoder_h1, encoder_b1, encoder_h2, encoder_b2):
    layer_1 = tf.nn.sigmoid(tf.add(tf.matmul(X, encoder_h1), encoder_b1))
    layer_2 = tf.nn.sigmoid(tf.add(tf.matmul(layer_1,encoder_h2), encoder_b2))
    
    return layer_2
```

```python
def decoder(X, decoder_h1, decoder_b1, decoder_h2, decoder_b2):
    layer_1 = tf.nn.sigmoid(tf.add(tf.matmul(X, decoder_h1), decoder_b1))
    layer_2 = tf.nn.sigmoid(tf.add(tf.matmul(layer_1,decoder_h2), decoder_b2))
    
    return layer_2
```

```python
X = tf.placeholder(tf.float32, [None, number_features])
encoder_model = encoder(X,encoder_h1, encoder_b1, encoder_h2, encoder_b2)
decoder_model = decoder(encoder_model,decoder_h1, decoder_b1, decoder_h2, decoder_b2)
```

```python
y_pred = decoder_model
y_true = X
cost = tf.reduce_mean(tf.pow(y_true - y_pred, 2))
optimizer = tf.train.RMSPropOptimizer(0.001, 0.9).minimize(cost)
```

```python
with tf.Session() as sess:
    tf.global_variables_initializer().run()
    
    for i in range(10000):
        batch_x, batch_y = mnist.train.next_batch(64)
        _, l = sess.run([optimizer, cost], feed_dict={X: batch_x})
        if i % display_step == 0:
            print('Step %i: Minibatch Loss: %f' % (i, l))
    
    n = 4
    canvas_orig = np.empty((28 * n, 28 * n))
    canvas_recon = np.empty((28 * n, 28 * n))
    for i in range(n):
        batch_x, _ = mnist.test.next_batch(n)
        g = sess.run(decoder_model, feed_dict={X: batch_x})

        for j in range(n):
            canvas_orig[i * 28:(i + 1) * 28, j * 28:(j + 1) * 28] = \
                batch_x[j].reshape([28, 28])
        for j in range(n):
            canvas_recon[i * 28:(i + 1) * 28, j * 28:(j + 1) * 28] = \
                g[j].reshape([28, 28])

    print("Original Images")
    plt.figure(figsize=(n, n))
    plt.imshow(canvas_orig, origin="upper", cmap="gray")
    plt.show()

    print("Reconstructed Images")
    plt.figure(figsize=(n, n))
    plt.imshow(canvas_recon, origin="upper", cmap="gray")
    plt.show()
```
